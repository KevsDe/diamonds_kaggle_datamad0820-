```python
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import keras.backend as kb
import tensorflow as tf
```

```python
predictors = pd.read_csv('X.csv',index_col=0)
target = pd.read_csv('Y.csv',index_col=0)
pred = pd.read_csv('clean_pred.csv',index_col=0)
train = pd.read_csv('input/train.csv', index_col=0)
```

```python
predictors.head()
```

```python
model = Sequential()
n_cols = predictors.shape[1]
model.add(Dense(5, activation='relu', input_shape=(n_cols,)))
model.add(Dense(5, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
batch_size = 70 
epochs = 80

history = model.fit(predictors, target,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(predictors, target))
```

```python
pred_train= model.predict(predictors)
print(np.sqrt(mean_squared_error(target,pred_train)))
print((r2_score(target,pred_train)))
```

```python
print(history.history.keys())
# "Loss"
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```
